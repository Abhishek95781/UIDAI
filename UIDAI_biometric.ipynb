```python
import pandas as pd
bio_1 = pd.read_csv('/content/api_data_aadhar_biometric_0_500000.csv')
bio_2 = pd.read_csv('/content/api_data_aadhar_biometric_1000000_1500000.csv')
bio_3 = pd.read_csv('/content/api_data_aadhar_biometric_1500000_1861108.csv')
bio_4 = pd.read_csv('/content/api_data_aadhar_biometric_500000_1000000.csv')
```

**MERGE **

```python
biometric = pd.concat(
    [bio_1, bio_2, bio_3, bio_4],
    ignore_index=True
)
```

**CONVERT DATE COLUMN**

```python
biometric['date'] = pd.to_datetime(
    biometric['date'],
    dayfirst=True,
    errors='coerce'
)
```

**Create total biometric count**

```python
biometric['total_biometric'] = (
    biometric['bio_age_5_17'] + biometric['bio_age_17_']
)
```

**GROUPBY (Univariate)**

```python
biometric.groupby('state')['total_biometric'].sum()
```

**Groupby (BIVARIATE)**

```python
biometric.groupby(
    ['state','date']
)['total_biometric'].sum()
```

**Groupby (Multivariate)**

```python
bio_grouped = biometric.groupby(
    ['state','district','pincode','date'],
    as_index=False
)['total_biometric'].sum()
bio_grouped.head()
```

**Save grouped output **

```python
bio_grouped.to_csv('biometric_grouped.csv', index=False)
```
